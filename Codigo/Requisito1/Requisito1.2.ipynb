Importa los módulos necesarios

```python
import os #Manejo de rutas y directorios.
import glob #Búsqueda de archivos con patrones (*.ris).
```

Fusionar múltiples archivos .ris en un solo archivo consolidado llamado articulos_fusionados.ris.

```python
#Define las carpetas de origen y el archivo de salida
base_folder = "../../Bases_de_datos" #Carpeta base
subfolders = ["science_direct", "sage"] #Sub carpetas de las fuentes
output_file = "articulos_fusionados.ris"
# Abrir el archivo de salida en modo escritura
with open(output_file, "w", encoding="utf-8") as outfile: #Se abre merged_articles.ris en modo escritura (w), sobrescribiendo cualquier archivo previo.
    for folder in subfolders: #Recorre las subcarpetas y busca archivos .ris
        folder_path = os.path.join(base_folder, folder)  #Genera la ruta de cada subcarpeta (articles/ieee/, articles/science_direct/, articles/sage/).
        ris_files = glob.glob(os.path.join(folder_path, "*.ris"))  #Usa glob.glob("*.ris") para encontrar todos los archivos .ris dentro de cada carpeta.
        for ris_file in ris_files: #Lee cada archivo .ris y lo copia en el archivo final
            with open(ris_file, "r", encoding="utf-8") as infile: #Abre cada archivo .ris en modo lectura (r).
                outfile.write(infile.read())  #Escribir el contenido en el archivo de salida
                outfile.write("\n\n")  #Agregar un espacio entre archivos para mejor lectura
print(f"Archivos combinados exitosamente en {output_file}")
```
